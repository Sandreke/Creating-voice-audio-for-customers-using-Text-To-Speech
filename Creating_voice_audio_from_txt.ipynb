```python
from gtts import gTTS         # Para convertir text-to-speech (TTS)
import os                     # Para manipular el sistema operativo
import pandas as pd           # Para manipular tablas
import time                   # Para mantener en pausa durante cierto periodo de tiempo
import math                   # Para el redondeo de números
from mutagen.mp3 import MP3   # Para manipular audios MP3, en este caso, saber la duración de los audios
```

### Paso 1: Crear los directorios para el input (txt) y el ouput (mp3)

```python
# Crear los directorios (carpetas)
directorios =['Archivos', 'Audios'] # Archivos txt con los mensajes

for i in directorios:
    os.makedirs(i)  

# # Por si se quiere eliminar algún directorio (incluso con archivos dentro de este)
# import shutil
# shutil.rmtree('Archivos/')
```

### Paso 2: Leer base de datos

```python
data = pd.read_excel('Datos.xlsx', sheet_name='Ventas')
data
```

### Pasos 3, 4, 5, 6 y 7

```python
for i in range(len(data)):
    
    persona = data.loc[i,'Nombre']
    producto = data.loc[i,'Producto']
    
    # Paso 3: Crear un archivo txt para cada persona
    ruta = 'Archivos/' + persona + '.txt'
    text_file = open(ruta, "w")
                     
    # Paso 4: Crear el mensaje
    msj_audio = 'Hola ' + persona + ', gracias por comprar ' + producto + ' con nosotros.'
    
    # Paso 5: Escribir sobre el archivo txt con el mensaje personalizado y cerrarlo
    text_file.write(msj_audio)
    text_file.close() 

    # Paso 6: Leer el archivo txt
    f = open(ruta)
    txt = f.read()
    
    # Paso 7: Convertir texto a audio (Text-to-Speech) y guardar
    language = 'es'
    acento = 'com.mx'
    audio = gTTS(text = txt, lang = language, tld = acento, slow = False)
    audio.save('Audios/' + persona + '.mp3')
```

### Paso 8: Reproducir archivos de audio

```python
ruta_audios = 'Audios'
audios = [audio for audio in os.listdir(ruta_audios) if os.path.isfile(os.path.join(ruta_audios, audio))]

for i in audios:
    ruta = os.path.join(ruta_audios, i)
    os.system(ruta)
    duracion = MP3(ruta).info.length
    time.sleep(math.ceil(duracion)) # Se espera un tiempo redondeo hacia arriba de la duración de cada audio
```
